```python
acids = data["ACID"].unique()
```

```python
updated_vol = []
```

```python
for index, row in data.iterrows():
    if row["ACID"] == 'HCl':
        v_acid = row["Acid_vol"]
        v_h2o = row["H2O_vol"]
        v_hcl = v_acid*0.36
        v_h2o_hcl = v_acid - v_hcl
        updated_vol.append(v_h2o +v_h2o_hcl)
        data.loc[index,['H2O_vol']] = v_h2o + v_h2o_hcl
        
    
    
```
