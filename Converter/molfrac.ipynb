```python
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_excel('DATA_pKa_updated.xlsx')
data
```

```python
mw = {
    "HCl":36.5,
    "H2SO4":98.078,
    "AA":60.052,
    "HNO3":63.01,
    "TFAA":114.02,
    "DClAA":128.94,
    "ClAA":94.5,
    "H2O":18,
    "DMF":73.09,
    "DMA":45.0837,
    "FA":46.03
}
ro = {"HCl":1.2,
    "H2SO4":1.83,
    "AA":1.05,
    "HNO3":1.51,
    "TFAA":1.49,
    "DClAA":1.56,
    "ClAA":1.58,
    "H2O":0.99669, 
    "DMF":0.944,
    "DMA":0.89
     }
```

```python
def n(ro, vol, mw):
    return (ro*vol)/mw
def n_frac(n_comp, n_tot):
    return n_comp/n_tot

def n_hcl(v, mw, c=12):
    return (c*v)/mw

def n_hno3(v, mw, c=15.8):
    return (c*v)/mw
```

```python
nfDMF = []
nfH2O = []
nfAcid = []
```

```python
for index, row in data.iterrows():
    n_DMF = n(ro['DMF'], row['DMF'], mw['DMF'])
    n_H2O = n(ro['H2O'], row['H2O_vol'], mw['H2O'])
    n_acid = row['Acid_moles']
    
    '''
    if row['ACID'] != 'HCl' or row['ACID']!= 'HNO3':
        n_acid = n(ro[row['ACID']], row['Acid_vol'], mw[row['ACID']])
    elif row['ACID'] == 'HCl':
        n_acid = n_hcl(row['Acid_vol'], mw[row['HCl']])
    elif row['ACID'] == 'HNO3':
        n_acid = n_hcl(row['Acid_vol'], mw[row['HNO3']])
    '''
        
    n_tot = n_DMF + n_H2O + n_acid
    
    nfDMF.append(n_frac(n_DMF, n_tot))
    nfH2O.append(n_frac(n_H2O, n_tot))
    nfAcid.append(n_frac(n_acid, n_tot))
    
    
```

```python
data['nf DMF'] = nfDMF
data['nf H2O'] = nfH2O
data['nf ACID'] = nfAcid
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
data
```

```python
data.to_excel('DATA_frac_moles.xlsx')
```
