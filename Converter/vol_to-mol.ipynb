```python
data = pd.read_excel("DATA_moles.xlsx")
```

```python
def moles(ro, v, mr):
    return (ro*v)/mr

def moles_hcl(v):
    return 12*(v/1000)
def moles_hno3(v):
    return 15.8*(v/1000)
```

```python
mr ={'TFAA':114.023,
    'AA':60,
    'DClAA':128.94,
    'H2SO4': 98.079 ,
    'HNO3':63,
    'ClAA':94.5}
ro ={'TFAA':1.49,
    'AA':1.05,
    'DClAA':1.5634,
    'H2SO4':1.83,
    'HNO3':1.51,
    'ClAA':1.58}
```

```python
acid_moles = []
```

```python
for index, row in data.iterrows():
    if row["ACID"] != 'HCl' and row["ACID"] != 'HNO3' and row["ACID"] != 'ClAA':
        if row["Acid_vol"] != 'nan':
            mls = moles(ro[row["ACID"]], row["Acid_vol"], mr[row["ACID"]])
            acid_moles.append(mls)
    elif row["ACID"] == 'HCl' :
        '''
        Because HCl is Aq. splution 
        '''
        mls = moles_hcl(row["Acid_vol"])
        acid_moles.append(mls)
        
    elif row["ACID"] == 'HNO3' :
        mls = moles_hno3(row["Acid_vol"])
        acid_moles.append(mls)
     
    elif row["ACID"] == 'ClAA' :
        acid_moles.append(row['Acid_moles'])
    
    
    
    
    
    
    
    
```

```python
acid_moles
```

```python
data["Acid_moles"] = acid_moles
```

```python
data.to_excel("DATA_moles.xlsx")
```
