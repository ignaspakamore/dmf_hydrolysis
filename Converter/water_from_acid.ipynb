```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("DATA_moles.xlsx")
data['H2O_vol'] = data['H2O_vol'].round(2)
data['Acid_vol'] = data['Acid_vol'].round(2)
```

```python
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
data
```

```python
acids = data["ACID"].unique()
```

```python
updated_vol = []
```

```python
for index, row in data.iterrows():
    if row["ACID"] == 'HCl':
        v_acid = row["Acid_vol"]
        v_h2o = row["H2O_vol"]
        v_hcl = v_acid*0.36
        v_h2o_hcl = v_acid - v_hcl
        updated_vol.append(v_h2o +v_h2o_hcl)
        data.loc[index,['H2O_vol']] = v_h2o + v_h2o_hcl
    elif row["ACID"] == 'HNO3':
        v_acid = row["Acid_vol"]
        v_h2o = row["H2O_vol"]
        v_hno3 = v_acid*0.7
        v_h2o_hno3 = v_acid - v_hno3
        updated_vol.append(v_h2o +v_h2o_hno3)
        data.loc[index,['H2O_vol']] = v_h2o + v_h2o_hcl
           
        
    
    
```
